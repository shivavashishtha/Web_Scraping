# Initialising the URL

```python
url="https://www.bbc.co.uk/asiannetwork/vote/top-songs/"
```

# Importing the Libraries

```python
import requests 
from bs4 import BeautifulSoup
```

# Creating a response

```python
res=requests.get(url)
```

```python
res
```

```python
test=res.text
```

# Creating a Soup

```python
soup=BeautifulSoup(test,"html.parser")
```

```python
print(soup.prettify())
```

```python
soup.title
```

# Scraping the Song names

```python
songs=soup.find_all(name="p",class_="intro")
```

```python
songs
```

# Scraping the Text of Songs

```python
titles=[song.getText() for song in songs]
```

```python
titles
```

# Creating a Dataframe

```python
df=pd.DataFrame(titles,columns=["song"])
```

```python
df = df.iloc[2:-2 , :]
```

```python
df.song=df.song.str.split(":")
```

```python
df.head()
```

```python
df
```

```python
df=pd.DataFrame(df['song'].to_list(),columns=["test","song","test1"])
```

```python
df
```

```python
df.drop(labels=["test","test1"],axis=1,inplace=True)
```

```python
df
```

# Exporting as CSV File

```python
df.to_csv("top100songs.csv")
```
